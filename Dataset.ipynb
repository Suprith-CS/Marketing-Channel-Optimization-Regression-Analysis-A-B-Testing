```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Create date range for 2025
dates = pd.date_range(start="2025-01-01", end="2025-12-31")

data = {
    'Date': dates,
    # Facebook Data: Higher conversion logic
    'FB_Views': np.random.randint(800, 1500, size=len(dates)),
    'FB_Clicks': np.random.randint(40, 100, size=len(dates)),
    'FB_Conversions': np.random.randint(8, 20, size=len(dates)),
    'FB_Cost': np.round(np.random.uniform(20, 50, size=len(dates)), 2),
    
    # Adwords Data: Higher views but lower conversion logic
    'Adwords_Views': np.random.randint(1000, 2000, size=len(dates)),
    'Adwords_Clicks': np.random.randint(30, 80, size=len(dates)),
    'Adwords_Conversions': np.random.randint(2, 10, size=len(dates)),
    'Adwords_Cost': np.round(np.random.uniform(25, 60, size=len(dates)), 2)
}

df = pd.DataFrame(data)

# Adding calculated columns as seen in the video
df['FB_CTR (%)'] = (df['FB_Clicks'] / df['FB_Views'] * 100).round(2)
df['FB_Conv_Rate (%)'] = (df['FB_Conversions'] / df['FB_Clicks'] * 100).round(2)
df['Adwords_CTR (%)'] = (df['Adwords_Clicks'] / df['Adwords_Views'] * 100).round(2)
df['Adwords_Conv_Rate (%)'] = (df['Adwords_Conversions'] / df['Adwords_Clicks'] * 100).round(2)

# Save to CSV
df.to_csv('marketing_campaign_2025.csv', index=False)
print("Dataset created successfully: marketing_campaign_2025.csv")
```
